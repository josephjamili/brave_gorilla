# Storytelling Data Visualization on Exchange Rates

The dataset we're using describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The dataset can be found on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020), where the information was sourced from the European Centrak Bank. 

Note that this dataset was downloaded on January 2021, as the dataset does get updated regularly.

````python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

exchange_rates.head()
````

````python
exchange_rates.info()
````

We see there are 5699 rows and 41 columns in the dataset. Most columns are listed as object, except for ```Iceland krona```, ```Romainan leu```, and ```Turkish lira```.

### Data Cleaning

We will need to clean the data first before we can create data visualizations. Much of the data cleaning code will have been provided, as the purpose of this project will be to work on coding for visualizations.

Our focus for the data will be on the exchange rates between the Euro and the American dollar.

Below, we do the following:

* We rename the ```[US dollar ]``` and ```Period\Unit:``` columns to something easier to type — ```US_dollar``` and ```Time```.
* We change the ```Time``` column to a ```datetime``` data type.
* We sort the values by ```Time``` in ascending order.
* We reset the index (and drop the initial index).

````python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
````

````python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
# Isolating roles from initial Dataframe
euro_to_dollar['US_dollar'].value_counts() 
# 62 counts of '-' which need to be droppped
````

````python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
# We drop the rows containing '-' and convert the column to float data type 

euro_to_dollar.info()

# 5637 rows x 2 columns
````

### Rolling Mean

We have finished cleaning the data. Now we will generate a line plot to visualize the evolution of the euro-US dollar exchange rate.

````python
import matplotlib.pyplot as plt
%matplotlib inline 
# Enables Jupyter to display graphs

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
````

We will calculate the rolling mean for the US_dollar column, using a moving window of 30 days.

````python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
````

````python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean']) # Moving window at 30 days
plt.show()
````

We're going to add a visual story highlighting how the euro-dollar rate had changed with each US Presidency between the years 2000 - 2020. To do that, we will highlight parts of the plot generated above to reflect the main color of their political party.

We will do this as a three separate plots, which each plot highlighting the period of time that George W. Bush, Barack Obama, and Donald Trump were in office between the time period mentioned. We will then use a fourth plot to show the overall connection between the three plots.

````python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001)
                                        & (euro_to_dollar['Time'].dt.year < 2021)]

bush = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year < 2009)]
obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009)
                             & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017)
                             & (bush_obama_trump['Time'].dt.year < 2021)]
````

````python
import matplotlib.style as style
style.use('fivethirtyeight') # using fivethrityeight style

## Adding subplots
plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

## Changes to all subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'],
                      alpha=0.3)
    ax.grid(alpha=0.5)
    
## Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'],
        color='#af0b1e')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                    '2007', '', '2009'],
                   alpha=0.3)
ax1.text(731516.0, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#af0b1e')
ax1.text(731516.0, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)

## Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'])
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                    '2015', '', '2017'],
                   alpha=0.3)
ax2.text(734288.0, 1.92, 'OBAMA', fontsize=18, weight='bold')
ax2.text(734288.0, 1.8, '(2009-2017)', weight='bold',
        alpha=0.3)

## Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'],
        color='#af0b1e')
ax3.set_xticklabels(['', '2017', '', '2018', '', '2019', '',
                    '2020', '', '2021'],
                   alpha=0.3)
ax3.text(736855.0, 1.92, 'TRUMP', fontsize=18, weight='bold')
ax3.text(736855.0, 1.8, '(2009-2017)', weight='bold',
        alpha=0.3)

## Ax4: Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'],
       color='#af0b1e')
ax4.plot(obama['Time'], obama['rolling_mean'])
ax4.plot(trump['Time'], trump['rolling_mean'],
       color='#af0b1e')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(730016.0, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

plt.show()
````
